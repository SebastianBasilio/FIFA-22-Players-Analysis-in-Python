## 1. DATA CLEANING

### 1.1 Manejo de NaN

```python
df_short.isnull().sum()
```

```python
df_short[df_short["club_name"].isna()].head(15)
```

```python
#Se eliminan las filas con muchos NaN
df_short=df_short.dropna(axis='index', how='any', subset=['club_name'])
#df_short.isnull().sum()
```

#### Los valores NaN de value_eur se van a llenar con valores comunes media/mediana acorde al overall de los jugadores

```python
#examino los NaN de value_eur 
df_short[df_short["value_eur"].isna()].head(13)
```

```python
df_short_overall=df_short[(df_short["overall"]>=55) & (df_short["overall"]<=75)]
#df_short_overall.tail(10)

# gráfico de distribución
x = df_short_overall['value_eur'].values

sns.displot(x, color = 'blue')

# gráfico de media
mean = df_short_overall['value_eur'].mean()

plt.axvline(mean, 0,1, color = 'red')

#tiene cola a la derecha, se llenarán los malores con la mediana
```

```python
#Se llenan los valores nan de value_eur con la mediana acorde al overall (puntuación de jugador)
df_short["value_eur"].fillna(df_short.groupby("overall")["value_eur"].transform("median"),inplace=True)

#df_short.isnull().sum()
```

```python
#Se comprueba el llenado adecuado
df_short[(df_short['sofifa_id']==138703) | (df_short['sofifa_id']==152912)]
```

#### Los NaN de la columna club joined se mantienen; ya que no afectan al análisis realizado

### 1.2 Manejo de Duplicados

```python
print(df_short.shape)
# Valores únicos de cada columna
for column in df_short:
    unique_vals= np.unique(df_short[column])
    nr_values = len(unique_vals)
    if nr_values <10:
        print('The number of values for feature {} : {} -- {}'.format(column,nr_values,unique_vals))
    else:
        print('The number of values for feature {} : {}'.format(column,nr_values))
```

```python
df_duplicated = df_short[df_short.duplicated(subset=['long_name'], keep=False)]
df_duplicated.sort_values(by="long_name").head(30)
```

#### 2.1 Relación entre Potential, overall, wage, value, height and age

```python
data = df_short[['potential','overall','wage_eur', 'value_eur', 'height_cm', 'age']].corr(method ='pearson')

cols = ['potential','overall','wage_eur', 'value_eur', 'height_cm', 'age']

plt.figure(figsize=(15,7))
ax = sns.heatmap(data, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30})

plt.title("Correlaciones", fontsize = 18)

#Se observa que el overall influye mas que el potential en el salario y valor
```

```python
ax = sns.pairplot(df_short[['potential','overall','wage_eur', 'value_eur', 'height_cm']], plot_kws={'color':'#0EDCA9'})
```

##### como se observo previamente en la matriz de correlación, los resultados coinciden

```python
# Comparo overall vs wage_eur añadiendo una división por la reputación internacional del jugador
plt.figure(figsize=(12, 7))
ax = sns.scatterplot(x ='overall', y = 'wage_eur', data=df_short,hue = 'international_reputation')
plt.xlabel("Overall") 
plt.ylabel("Wage EUR")
plt.title("Overall & wage", fontsize = 14)
plt.show()

#Se obseva que hay una fuerte relación entre overall y reputación, y que este último no afecta en 
# gran medida al salior (a excepción de los jugadores más pagados) 
```

#### 2.5 Cual es la posición más común en el top 100 (acorde al overall) de jugadores


```python
def plot_most_common(category):
    df_100 = df_short.nlargest(100, 'overall')
    bar_plot = dict(Counter(df_100[category].values).most_common(5))
    plt.bar(*zip(*bar_plot.items()))
    plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,5))
plot_most_common('club_position')
plt.show()
```
